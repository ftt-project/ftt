# Hedging

It is the only effective way to obtain downside protection

Dynamic hedging

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import pandas as pd
import numpy as np
import os,sys,inspect
current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parent_dir = os.path.dirname(current_dir)
sys.path.insert(0, parent_dir) 
import edhec_risk_kit as erk
```

```python
ind_return = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

```python
ind_return.shape
```

```python
ind_nfirms.shape
```

```python
ind_size.shape
```

```python
ind_nfirms.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
total_mktcap = ind_mktcap.sum(axis="columns")
```

ind_mktcap - market value of particular industry

total_mktcap - overall market value at the given point of time

```python
total_mktcap.plot(figsize=(12, 6))
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis="rows")
ind_capweight.head()
```

ind_capweight - is a fraction, that industry takes in a market

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin", "Steel"]].plot(figsize=(12,6))
```

```python
total_market_return = (ind_capweight * ind_return).sum(axis="columns")
```

The total Market return is the weighted average of each one of these industries. Take the capweight, and you multiply by the return of that industry.

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(10,5), title="Total Market CapWeighted Index 1926-2018")
```

total_market_index - is like an index of the stock, but of all market

Wealth is useful, it shows the wealthy of the stock

```python
total_market_index["1980":].plot(figsize=(12,6))
# Moving average
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6))
```

## Rolling Correlation - along with MultiIndexes and `.groupby`

```python
# time series correlation
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names = ["date", "industry"]
```

```python
ts_corr.tail()
```

```python
 ind_tr36corr = ts_corr.groupby(level="date").apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets", legend=True, figsize=(12,6))
ind_tr36corr.plot(label="Tr36 Mo Corr", legend=True, figsize=(12,6))
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets", legend=True, figsize=(12,6))
ind_tr36corr.plot(label="Tr36 Mo Corr", legend=True, figsize=(12,6), secondary_y=True)
```

### Only 2007 to zoom in market crash in 2008

```python
tmi_tr36rets["2007":].plot(label="Tr36 Mo Rets", legend=True, figsize=(12,6))
ind_tr36corr["2007":].plot(label="Tr36 Mo Corr", legend=True, figsize=(12,6), secondary_y=True)
```
