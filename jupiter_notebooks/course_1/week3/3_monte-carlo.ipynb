```python
import numpy as np
import pandas as pd

def gbm(n_years=10, n_scenarios=100, mu=.07, sigma=.15, steps_per_year=12, s_0=100.0):
    """
    Evolution of a Stock Price using Geometric Brownian Motion model
    """
    dt = 1/steps_per_year
    n_steps = int(n_years * steps_per_year)
    # xi = np.random.normal(size=(n_steps, n_scenarios))
    rets = np.random.normal(loc=1+mu*dt, scale=(sigma * np.sqrt(dt)),size=(n_steps+1, n_scenarios))
    # rets = mu * dt + sigma * np.sqrt(dt) * xi
    # rets = pd.DataFrame(rets)
    # rets to prices
    # prices = s_0 * (1 + rets).cumprod()
    rets[0] = 1 # to start from 100 as a start point
    prices = s_0 * pd.DataFrame(rets).cumprod()
    return prices
```

```python
p = gbm(10, n_scenarios=3)
```
